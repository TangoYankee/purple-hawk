```python
import pandas as pd
import geopandas as gpd
```

```python
base_path = "../../purple-hawk-data"
source_path = f"{base_path}/source"
dest_path = f"{base_path}/db-copy"
```

```python
facilities_full = gpd.read_file(f"{source_path}/facilities_23v2_shp")
```

```python
facilities_lift = facilities_full.reindex(
    [
        "UID",
        "FACNAME",
        "BIN",
        "BBL",
        "ADDRESSNUM",
        "STREETNAME",
        "FACTYPE",
        "SERVAREA",
        "OPNAME",
        "OVERAGENCY",
        "CAPACITY",
        "CAPTYPE",
        "geometry",
    ],
    axis=1,
)
```

```python
facilities_wgs84 = facilities_lift.to_crs(4326)
```

```python
facility = pd.DataFrame(data=facilities_lift)
facility.drop("geometry", axis=1, inplace=True)
facility["lift"] = facilities_lift["geometry"]
facility["wgs84"] = facilities_wgs84["geometry"]
```

```python
facility["BBL"] = facility["BBL"].astype(int).astype(str)
```

```python
facility.to_csv(f"{dest_path}/facility_v1.csv", index=False)
```
